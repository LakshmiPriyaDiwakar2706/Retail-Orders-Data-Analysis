```python
#replacing duplicate values with NaN
df = pd.read_csv(r"C:\Users\laksh\OneDrive\Desktop\Data Science\Python+SQL\retail-orders\orders.csv",na_values=['Not Available','unknown'])
```

```python
#renaming columns
#Convert the columns into string object

df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
#derive new columns dicount, sales price, profit

df['discount'] = df['list_price']*df['discount_percent']*0.01
df['sales_price']=df['list_price']-df['discount']
df['profit'] = df['sales_price']-df['cost_price']
```

```python
df.head()
```

```python
#convert order date to datetime datatype

df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df.info()
```

```python
#drop cost price, list price and discount percent as it is not needed right now

df = df.drop(columns=['list_price','cost_price','discount_percent'])

```

```python
#connecting to PostgreSQL database
#importing libraries
from sqlalchemy import create_engine
```

```python
#defining PostgreSQL connection

username = 'postgres'
password = '-----'
hostname = 'localhost'
port = 5432
database = 'analysis'
connection_string = f'postgresql+psycopg2://{username}:{password}@{hostname}:{port}/{database}'

#create SQLAlchemy engine
engine = create_engine(connection_string,pool_pre_ping=True)

```

```python
#Export DataFrame to PosgreSQL

df.to_sql('retail_orders',engine,if_exists='replace',index=False)
print("DataFrame exported to PostgreSQL successfully!")
```
